```python
#importando cplex
from docplex.mp.model import Model
```

```python
#Inicializando el modelo
mdl=Model('Transmision')
```

```python
#Creando las variables de decision
#x=mdl.integer_var_dict(tramos,name='x')
x=mdl.continuous_var_dict(tramos,name='x')
```

```python
#Funcion objetivo de minimizacion y formacion del modelo
mdl.minimize(mdl.sum(x[i]*Costos[i] for i in tramos))
```

```python
#Restriccion correspondiente a la oferta por ciudad
for k in ciudadesi:
    mdl.add_constraint(mdl.sum(x[k,j]for j in ciudadesd if j!=k)<=Oferta[k])
    #mdl.add_constraint(mdl.sum(x[k,j]for j in ciudadesd)<=Oferta[k])
```

```python
#Restriccion correspondiente a la demanda por ciudad
for k in ciudadesd:
    mdl.add_constraint(mdl.sum(x[k,j]for j in ciudadesi if j!=k)>=Demanda[k])
    #mdl.add_constraint(mdl.sum(x[k,j]for j in ciudadesi)>=Demanda[k])
```

```python
#Modelo (Funcion objetivo, restricciones y tramos)
print(mdl.export_to_string())
```

```python
solucion=mdl.solve(log_output=True)
```

```python
#restricciones del modelo
n_const=mdl.number_of_constraints
const=[mdl.get_constraint_by_index(i) for i in range(n_const)]
const
```

```python
#variables de holgura
h=mdl.slack_values(const)
h
```

```python
#Variables de holgura de cada una de las restricciones
for n in range (n_const):
    print("La variable de holgura de la restriccion "+str(const[n])+" es "+str(h[n]))
```

```python
#Precios duales

precio_dual=mdl.dual_values(const)

#Precio dual para cada restriccion

for n in range (n_const):
    print("El precio dual de la restriccion "+str(const[n])+" es "+str(precio_dual[n]))
```

```python
#sensibilidad
cpx=mdl.get_engine().get_cplex()
#coeficiente de la funcion objetivo
of=cpx.solution.sensitivity.objective()
#calculo del lado derecho
b=cpx.solution.sensitivity.rhs()
b
```
