```python
from docplex.mp.model import Model
```

```python
mdl=Model('Transmision')
```

```python
x=mdl.integer_var_dict(tramos,name='x')
```

```python
mdl.minimize(mdl.sum(x[i]*Costos[i] for i in tramos))
```

```python
for k in ciudadesi:
    mdl.add_constraint(mdl.sum(x[k,j]for j in ciudadesd if j!=k)<=Oferta[k])
```

```python
for k in ciudadesd:
    mdl.add_constraint(mdl.sum(x[k,j]for j in ciudadesi if j!=k)>=Demanda[k])
```

```python
print(mdl.export_to_string())
```

```python
solucion=mdl.solve(log_output=True)
```

```python
#variables de holgura
h=mdl.slack_values(const)
h
```
